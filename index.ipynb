```python
# Let's get started importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the data
df = pd.read_csv('ames_train.csv')
df.head(20)
```

```python
# Investigate the Data
df.info()
df.shape
df.columns
```

```python
# Investigating Distributions using scatter_matrix
pd.plotting.scatter_matrix(df[['YearBuilt','LotArea','SalePrice']], figsize=(10,10))
```

```python
# Create a plot that shows the SalesPrice Distribution
ax = df['SalePrice'].plot.hist(bins=20, figsize = (10,10))
ax.set_xlabel('Price')
ax.set_title('Sales Price Distribuition')
ax.set_ylabel('Number of Houses')
ax.axvline(df['SalePrice'].mean(), color='red');
ax.axvline(df['SalePrice'].median(), color = 'black')
```

```python
# Create a plot that shows the LotArea Distribution
ax_2 = df['LotArea'].plot.hist(bins=200, figsize = (10,10))
ax_2.set_xlabel('Lot Area')
ax_2.set_title('Lot Area Distribuition')
ax_2.set_ylabel('Number of Houses')
ax_2.axvline(df['LotArea'].mean(), color='red');
ax_2.axvline(df['LotArea'].median(), color = 'black')
```

```python
# Create a plot that shows the Distribution of the overall house condition
ax_3 = df['OverallCond'].plot.hist(figsize = (10,10))
ax_3.set_title('Overall House Condition Distribuition')
ax_3.set_ylabel('Number of Houses')
ax_3.set_xlabel('House Condition Scale')
ax_3.axvline(df['OverallCond'].mean(), color='red');
ax_3.axvline(df['OverallCond'].median(), color = 'black')
```

```python
# Create a Box Plot for SalePrice
ax_4 = df['SalePrice'].plot.box(figsize = (10,10))
ax_4.set_ylabel('Sale Price $')
ax_4.set_xlabel(' x ')
```

```python
# Perform an Exploration of home values by age
df['Age'] = 2020 - df['YearBuilt']
df['Decades'] = df['Age'] // 10
ax_5 = df.groupby('Decades').SalePrice.mean()
ax_6 = ax_5.plot.bar(figsize = (10,10))
ax_6.set_title('Home Values By Decades')
ax_6.set_ylabel('Price ($)')
ax_6.set_xlabel('Age by Decade')
```
